# SPLIT

```python
X=df.iloc[1:,0:13].values
Y=df.iloc[1:,13:].values
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Random forest 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay,f1_score
```

```python
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
rf.fit(x_train,y_train);
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf=precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_score_rf=f1_score(y_test, y_pred_rf)
```

```python
print("Accuracy:",accuracy_rf)
print("Precision:",precision_rf)
print("Recall:",recall_rf)
print("f1_score:",f1_score_rf)
```

```python
print(confusion_matrix(y_test, y_pred_rf))
```

# Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
lg=LogisticRegression()
lg.fit(x_train,y_train)
y_pred_lg=lg.predict(x_test)
```

```python
accuracy_lg = accuracy_score(y_test, y_pred_lg)
precision_lg=precision_score(y_test, y_pred_lg)
recall_lg = recall_score(y_test, y_pred_lg)
f1_score_lg=f1_score(y_test, y_pred_lg)
```

```python
print("Accuracy:",accuracy_lg)
print("Precision:",precision_lg)
print("Recall:",recall_lg)
print("f1_score:",f1_score_lg)
```

```python
print(confusion_matrix(y_test, y_pred_lg))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
training_accuracy_knn=[]
test_accuracy_knn=[]
neighbors=range(1,40)
for n in neighbors:
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(x_train,y_train)
    training_accuracy_knn.append(knn.score(x_train,y_train))
    test_accuracy_knn.append(knn.score(x_test,y_test))
```

```python
plt.plot(neighbors,training_accuracy_knn,label="training accuracy")
plt.plot(neighbors,test_accuracy_knn,label="test accuracy")
plt.xlabel("num of neighbors")
plt.ylabel("accuracy")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
y_pred_knn = knn.predict(x_test)
```

```python
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)
recall_knn=recall_score(y_test,y_pred_knn)
print('Accuracy:', accuracy_knn)
print('Precision:', precision_knn)
print('F1 score:', f1_knn)
print('Recall:',recall_knn)
```

```python
print(confusion_matrix(y_test,y_pred_knn))
```

# ROC and AUC

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import roc_auc_score
```

```python
probs=[0 for _ in range(len(y_test))]
rf_probability=rf.predict_proba(x_test)
lg_probability=lg.predict_proba(x_test)
knn_probability=knn.predict_proba(x_test)
svm_probability=svc.predict_proba(x_test)
```

```python
rf_probability=rf_probability[:,1]
lg_probability=lg_probability[:,1]
knn_probability=knn_probability[:,1]
svm_probability=svm_probability[:,1]
```

```python
rf_auc=roc_auc_score(y_test,rf_probability)
lg_auc=roc_auc_score(y_test,lg_probability)
knn_auc=roc_auc_score(y_test,knn_probability)
svm_auc=roc_auc_score(y_test,svm_probability)
```

```python
print("Random Forest: AUROC = % 0.3f" % (rf_auc))
print("Logistic Regression: AUROC = %0.3f" % (lg_auc))
print("KNN: AUROC = %0.3f" % (knn_auc))
print("Support Vector Machine: AUROC = %0.3f" % (svm_auc))
```

```python
rf_fpr,rf_tpr,_=roc_curve(y_test,rf_probability)
lg_fpr,lg_tpr,_=roc_curve(y_test,lg_probability)
knn_fpr,knn_tpr,_=roc_curve(y_test,knn_probability)
svm_fpr,svm_tpr,_=roc_curve(y_test,svm_probability)
```

```python
def plot_roc_curve(lr_fpr,lr_tpr):
    plt.plot(rf_fpr,rf_tpr,marker=".", label="Random Forest (AUROC = %0.3f)" %rf_auc)
    plt.plot(lg_fpr,lg_tpr,marker=".", label="Logistic Regression (AUROC = %0.3f)" %lg_auc)
    plt.plot(knn_fpr,knn_tpr,marker=".", label="KNN (AUROC = %0.3f)" %knn_auc)
    plt.plot(svm_fpr,svm_tpr,marker=".", label="SVM (AUROC = %0.3f)" %svm_auc)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Postive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.plot([0, 1], [0, 1])
    plt.show()
```

```python
plot_roc_curve(lg_fpr,lg_tpr)
```
